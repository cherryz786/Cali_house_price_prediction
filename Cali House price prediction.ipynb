```python
# importing all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data and reading csv file
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
# notice how the "total_bedrooms" have some values missing and they are not 20640
#so we simply drop those null values next
```

```python
data.dropna(inplace = True)
```

```python
# now check the new info 
# they all have 20433 non null values !
# what we have done here is dropped the entire columns where we have null values
data.info()
```

```python
# now we split the data into training and test data into x and y data
```

```python
# now we import sklearn to train our model

from sklearn.model_selection import train_test_split
```

```python
# now we have to split out data into training and test data in x and y
# here axis is 1 for x cause we are dropping an entire column

x = data.drop("median_house_value", axis=1)

y = data["median_house_value"]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

# here we are reserving 20 percent of data as test data
```

```python
train_data = x_train.join(y_train)

# here we are combining the training features (x_train) and the training target (y_train)
# to create a new DataFrame called train_data.



# The join method is a Pandas DataFrame method used to combine two DataFrames based on a common column or index. 
# In this case, the x_train DataFrame contains the training features, and the y_train DataFrame contains 
# the training target (the "median_house_value" column).



# When you use the join method, it aligns the rows of both DataFrames based on their index and combines them 
# into a single DataFrame. The resulting train_data DataFrame will have the columns from both x_train and y_train, 
# with matching rows based on their index.
```

```python
train_data.hist(figsize = (15,8))

# this gives multiple histograms based on various features of the data
# increase the figsize to avoid it overlapping one another
```

```python
train_data.corr()

# we are calculating the correlation matrix for the train_data DataFrame using the corr() method.


# The correlation matrix is a square matrix that shows the correlation coefficients between all pairs of 
# columns in a DataFrame. The correlation coefficient measures the strength and direction of a linear relationship 
# between two variables. It ranges from -1 to 1.
```

```python
# now we plot the corelation matrix

plt.figure(figsize=(15,8))

sns.heatmap(train_data.corr(), annot = True, cmap="YlGnBu")
```

```python
# here one thing to note in the above corelation matrix is that the median_income corelates highly with the median_house_value
# hence this can be an important factor in prediction by out model.
```

```python
# NOW WE DO THE DATA PREPROCESSING
```

```python
# This cell below applies the natural logarithm to the "total_rooms" column and other columns of the train_data DataFrame. 
# The + 1 is added before taking the logarithm to avoid the logarithm of zero or negative values. 
# The transformed values will replace the original values in the "total_rooms" column.
```

```python
train_data["total_rooms"] = np.log(train_data["total_rooms"] + 1)

train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] + 1)

train_data["population"] = np.log(train_data["population"] + 1)

train_data["households"] = np.log(train_data["households"] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.ocean_proximity.value_counts()
# so we have these types of ocea_proximity values, we can use this to estimate the price of the house cause this 
# feature is indeed significant
```

```python
pd.get_dummies(train_data.ocean_proximity)

# Dummy variables, also known as indicator variables, are a way to represent categorical data in a 
# binary format (0s and 1s). Each unique category in the original categorical variable becomes a new binary variable, 
# and the presence of a category is indicated by a 1, while the absence is indicated by a 0.
```

```python
train_data.join(pd.get_dummies(train_data.ocean_proximity))
```

```python
# now since we have joined the dummy vairables corresponding to ocean_procimity we no longer need it so we can drop this column
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(["ocean_proximity"], axis = 1) 

# axis is 1 cause we are dropping an entire column
```

```python
train_data
```

```python
# now we plot the corelation matrix again to see how these new features corelate with the values

plt.figure(figsize=(15,8))

sns.heatmap(train_data.corr(), annot = True, cmap="YlGnBu")
```

```python
# notice how "median_house_value" has negative corelation with "inland", which means if you buy a house inland you pay wayy less
```

```python
plt.figure(figsize=(15,8))

sns.scatterplot(x="latitude", y="longitude", data = train_data, hue = "median_house_value", palette = "coolwarm")
```

```python
# now we can visualise in the above scatter plot that all the houses in close proximity to the ocean has a red color which 
# indicates that they are highly priced
```

```python
# FEATURE ENGINEERING:
```

```python
train_data["bedroom_ratio"] = train_data["total_bedrooms"] / train_data["total_rooms"]

train_data["household_rooms"] = train_data["total_rooms"] / train_data["households"]
```

```python
# now we plot the corelation matrix again to see how these new features corelate with the values

plt.figure(figsize=(15,8))

sns.heatmap(train_data.corr(), annot = True, cmap="YlGnBu")
```

```python
# here we have enginnered two features which seems to be interesting and plays a major role in deciding the "median_house_value"
# these features are: "bedroom_ratio" and "household_rooms"
```

```python
# NOW WE TRAIN THE LINEAR REGRESSION MODEL:
```

```python
from sklearn.linear_model import LinearRegression

x_train, y_train = train_data.drop(["median_house_value"], axis = 1), train_data["median_house_value"]

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
# since "ocean_proximity" column has categories ["INLAND", "ISLAND", "NEAR BAY", "NEAR OCEAN"]
ocean_proximity_categories = ["INLAND", "ISLAND", "NEAR BAY", "NEAR OCEAN"]

# We combine the feature data x_test and the target data y_test into a single DataFrame called test_data. 
# This makes it easier to work with the test data as a single unit.
test_data = x_test.join(y_test)

# Applying Log Transformation:
test_data["total_rooms"] = np.log(test_data["total_rooms"] + 1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] + 1)
test_data["population"] = np.log(test_data["population"] + 1)
test_data["households"] = np.log(test_data["households"] + 1)

# Create dummy variables for the "ocean_proximity" column in test_data
dummy_cols = pd.get_dummies(test_data["ocean_proximity"], prefix="", prefix_sep="")
test_data = test_data.join(dummy_cols).drop(["ocean_proximity"], axis=1)

# Additional Feature Engineering:
test_data["bedroom_ratio"] = test_data["total_bedrooms"] / test_data["total_rooms"]
test_data["household_rooms"] = test_data["total_rooms"] / test_data["households"]
```

```python
x_test, y_test = test_data.drop(["median_house_value"], axis = 1), test_data["median_house_value"]
```
